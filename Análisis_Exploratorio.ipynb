```python
import pandas as pd
```

```python
# Leer el archivo CSV
df = pd.read_csv("master.csv", encoding='utf-8')
```

```python
# Imprimir cantidad de observaciones y variables
print(f"Número de filas: {len(df)}\nNúmero de columnas: {len(df.columns)}\n")

# Información general del DataFrame
print(df.info())

# Estadísticas descriptivas
print(df.describe())
```

```python
for column in df.columns:
    print(f"{column}: {df[column].describe()}")
    print(f"Presenta entradas con NAN: {df[column].hasnans}")
    if df[column].hasnans:
        print(f"Cantidad de observaciones con NAN: {df[column].isna().sum()}")
    print("--------------------------------------------------------------------")
```
